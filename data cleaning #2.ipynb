```python
import pandas as pd
import numpy as np

df = pd.read_csv("dataset_1.csv")
```

Hapus kolom yang tidak dibutuhkan

```python
hapus_kolom = ['country_new','chronic_disease','date_onset_symptoms','full_symptoms']
```

```python
df.drop(hapus_kolom, inplace=True, axis=1)
```

Hapus semua baris yang memiliki missing value

```python
df.isnull().sum()
```

```python
df.dropna(axis=0, inplace=True)
```

Hapus baris yang mengandung death, died, deceased (kolom outcome)

```python
df['outcome'].head(50)
```

```python
df['outcome'].tail(50)
```

```python
died = df[(df['outcome'] == 'died')].index
df.drop(died, inplace=True)
```

```python
died = df[(df['outcome'] == 'dead')].index
df.drop(died, inplace=True)
```

```python
death = df[(df['outcome'] == 'death')].index
df.drop(death, inplace=True)
```

```python
deceased = df[(df['outcome'] == 'Deceased')].index
df.drop(deceased, inplace=True)
```

Hapus kolom outcome, karena sudah dibersihkan dan tersisa tanggal pasien sembuh

```python
df.drop('outcome', inplace=True, axis=1)
```

```python
df
```

Rename nama kolom date_death_or_discharge menjadi date_recovered, karena sudah dibersihkan dan tersisa tanggal pasien sembuh

```python
df.rename(
    columns={'date_death_or_discharge' : 'date_recovered'}, 
    inplace = True,
)
```

Mengubah index sesuai default

```python
df.reset_index(drop=True, inplace=True)
```

```python
df
```

Mengubah value dari kolom sex menjadi 1 untuk male dan 0 untuk female

```python
def changetobiner(df):
    if df == 'male':
        df = 1
    else:
        df = 0  
    
    return df
  
sex = []
for index, row in df.iterrows():
    sex.append(changetobiner(row['sex']))

df['sex'] = sex
```

```python
df
```

Menambahkan kolom recovery_days

```python
df['date_confirmation'] = df['date_confirmation'].str.replace('.', '-')
df['date_recovered'] = df['date_recovered'].str.replace('.', '-')
```

```python
df['date_confirmation'] = pd.to_datetime(df['date_confirmation'], format='%d-%m-%Y')
df['date_recovered'] = pd.to_datetime(df['date_recovered'], format='%d-%m-%Y')

df['recovery_days'] = df['date_recovered'] - df['date_confirmation']
```
